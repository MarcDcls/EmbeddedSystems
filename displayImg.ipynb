```python
img = plt.imread('G:\Mon Drive\ENSC\S9\Mecatronique-systemeEmbarque\logo.jpg')
imggray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
(tresh, bw_img)=cv2.threshold(imggray,127,255,cv2.THRESH_BINARY)

value=np.sqrt((bw_img.shape[0]//2)**2+(bw_img.shape[1]//2)**2)

polarimg=cv2.linearPolar(bw_img,(bw_img.shape[0]/2,bw_img.shape[1]/2), value, cv2.WARP_FILL_OUTLIERS)

plt.imshow(polarimg)
plt.show()
```

```python
heigth=polarimg.shape[1]
width=polarimg.shape[0]
N=256
largeurCase=width//N
hauteurCase=heigth//18
# print(polarimg[1][1])
# print(polarimg[1])
polarimgTrans=np.transpose(polarimg)

imgplot = plt.imshow(polarimgTrans)

plt.show()
imgf=[]
for i in range(N):
    colonne=''
    for j in range(16):
        
        compt=0
        for k in range (largeurCase):
            for l in range(hauteurCase):
                compt=compt+polarimg[k+i*largeurCase+2*largeurCase][l+j*hauteurCase]
        moyenne=compt/(largeurCase*hauteurCase)
        if moyenne>127:
            colonne=colonne+'1'
        else:
            colonne=colonne+'0'
    imgf.append(colonne)
# imgf2=np.transpose(imgf)
# imgplot = plt.imshow(imgf)
# plt.show()
print(imgf)
```
